## Train and Save the Abalone Model

**Section Purpose:** Train the model once and save it (and feature names) so the Streamlit and Gradio apps can load it without retraining.

**What You'll Do:** Load the Abalone dataset (same as Lecture 09), apply the same preprocessing (`get_dummies` for Sex), build a pipeline (StandardScaler + LinearRegression), fit it, and save the pipeline and feature column list with `joblib`. Optionally report R² and RMSE.

**Success Criteria:** You have `abalone_model.joblib` and `feature_columns.joblib` in this folder, and you know the exact feature names and order the apps must use.

Place `abalone.csv` in this folder (from Canvas). We use the same preprocessing as Lecture 09.

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error
import joblib
````

````python
# Load Abalone data (same as Lecture 09)
abalone = pd.read_csv("abalone.csv", index_col=0)
abalone = pd.get_dummies(abalone, drop_first=True, prefix="Sex")
````

````python
X = abalone.drop("Rings", axis=1)
y = abalone[["Rings"]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
````

````python
# Pipeline: StandardScaler + LinearRegression (same as Lecture 09)
steps = [("scaler", StandardScaler()), ("lin_reg", LinearRegression())]
pipeline = Pipeline(steps)
pipeline.fit(X_train, y_train)
````

````python
pred = pipeline.predict(X_test)
r_squared = pipeline.score(X_test, y_test)
rmse = root_mean_squared_error(y_test, pred)
print(f"R^2: {r_squared}")
print(f"RMSE: {rmse}")
````

````python
# Save pipeline and feature names for the apps (save in current working directory)
model_path = "abalone_model.joblib"
features_path = "feature_columns.joblib"
joblib.dump(pipeline, model_path)
joblib.dump(X.columns.tolist(), features_path)
print(f"Saved pipeline to {model_path} and feature names to {features_path}")
print("Feature order the apps must use:", X.columns.tolist())
````

## Streamlit App (Local and Cloud)

**Section Purpose:** Build a simple web UI with Streamlit, test it locally, then deploy to Streamlit Community Cloud.

**What You'll Do:** Use the provided `app.py` (loads the model, defines inputs for each feature, predicts Rings). Test locally, then get Streamlit Cloud access and deploy.

**Success Criteria:** Run the app locally and get a correct prediction; deploy to Streamlit Community Cloud and open the live app.

### How to test locally

1. In the terminal, go to the `Lecture 15 - Deploying Models` folder.

### How to get Streamlit Community Cloud access

- **No separate API key** is required for basic deployment. You sign in with **GitHub**.

### How to deploy to Streamlit Community Cloud

1. Put your app in a **GitHub repository**. The repo (or a subfolder) must contain:
   - `app.py`
   - `requirements.txt`
   - `abalone_model.joblib` and `feature_columns.joblib` (so the cloud can load the model; run the notebook and commit these files, or add them from this folder).
2. In Streamlit Cloud, click **New app**.
3. Connect your GitHub account if needed, then choose the **repository**, **branch**, and **Main file path** (e.g. `app.py`).
4. If the app lives in a subfolder (e.g. `Lectures/Lecture 15 - Deploying Models/`), set **App root** to that folder so the cloud finds `app.py` and the model files.
5. Click **Deploy**. The cloud will build and host the app; you’ll get a public URL.

If the model files are missing in the repo, the deployed app will show: “Run the notebook to train and save the model first.”

## Gradio App (Local and Cloud)

**Section Purpose:** Build an alternative UI with Gradio, test it locally, then deploy to Gradio Cloud (Hugging Face Spaces).

**What You'll Do:** Use the provided `gradio_app.py` (same model and feature logic as the Streamlit app, exposed via Gradio). Test locally, then get a Hugging Face token and deploy to a Space.

**Success Criteria:** Run the Gradio UI locally and get a correct prediction; deploy to Hugging Face Spaces and open the live app.

### How to test locally

1. In the `Lecture 15 - Deploying Models` folder, run: `uv run python gradio_app.py`
2. Open the URL shown in the terminal (e.g. http://127.0.0.1:7860).
3. Enter inputs and submit; confirm the predicted Rings matches expectations (e.g. compare with the notebook or Streamlit app).

### How to get Gradio / Hugging Face access (API key)

- Gradio apps are typically deployed on **Hugging Face Spaces**. You need a **Hugging Face account** and an **access token**.
1. Create an account at [huggingface.co/join](https://huggingface.co/join) if you don’t have one.
2. Go to [huggingface.co/settings/tokens](https://huggingface.co/settings/tokens).
3. Click **New token**, give it a name (e.g. “Spaces deploy”), and choose **Write** (or at least write permission for Spaces) so you can push your app.
4. Copy and store the token securely; you’ll use it as the password when pushing to a Space via Git (or with the `huggingface_hub` CLI).

### How to deploy to Gradio Cloud (Hugging Face Spaces)

1. Go to [huggingface.co/new-space](https://huggingface.co/new-space).
2. Choose **Gradio** as the SDK, pick a Space name and visibility (public/private).
3. The Space needs at least:
   - `gradio_app.py` (the file that launches the Gradio app)
   - `requirements.txt`
   - `abalone_model.joblib` and `feature_columns.joblib`
4. Push your code and model files to the Space:
   - The Space page shows **Git** instructions. Use HTTPS and your **Hugging Face token** as the password when pushing.
   - Or clone the Space repo, add the files, then `git push`. On first push you may be prompted for username (your HF username) and password (your token).
5. After the push, Hugging Face will build and run the app. The live URL is: `https://huggingface.co/spaces/<your-username>/<space-name>`.

The model file must be in the Space (committed to the repo or uploaded) or the app will show an error.
